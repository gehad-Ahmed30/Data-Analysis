```python
import matplotlib.pyplot as plt  
import pandas as pd  
import numpy as np
```

```python
x = [1,2,4,5,6]
y = [3,4,8,15,12]
plt.scatter(x,y)
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]
plt.title("this is a linear graph")
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]
plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.xlabel("Time")
plt.ylabel("price")
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]

plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,label="y=2x",color='red')
plt.xlabel("Time")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]

plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,label="y=2x",color='red',linewidth=4)
plt.xlabel("Time")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]

plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,label="y=2x",color='red',marker='o')
plt.xlabel("Time")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]

plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,label="y=2x",color='red',marker='o',linestyle='--')
plt.xlabel("Time")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]
plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,label="y=2x",color='red',marker='o',linestyle=':')
plt.xlabel("Time")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]

plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,'b*--',label="y=2x")
plt.xlabel("Time")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = [0,1,2,4,5,6]
y = [0,2,4,8,10,12]
x2 = np.arange(0,4.5, 0.5) # 0 , 0.5,  1 , 1.5. 2,  2.5 , 3, 3.5, 4 
#x2 = [.5,1,1.5,2,2.5,3....]
plt.title("this is a linear graph",fontdict={"fontname":"Comic Sans MS",'fontsize':18})
plt.plot(x,y,'b*--',label="equation 1")
plt.plot(x2,x2**2,'g>--',label='equation 2')
plt.title("this is a good graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,6]
y = [2,4,8,10,12]
x2 = [0, 0.5 , 1 , 1.5 , 2, 2.5 ,3 , 3.5] # 0 , 0.5,  1 , 1.5. 2,  2.5 , 3, 3.5
plt.plot(x,y,'r*--' ,label = "equation 1")
plt.plot(x2,[(i*i)for i in x2],'g>--',label = "equation 2")
plt.title("this is a good graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```
