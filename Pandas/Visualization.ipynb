### 3D plots

```python
plt.rcParams['figure.figsize']=(8,6)
np.random.seed(31)
mu=3
n=50
x=np.random.normal(mu,1,size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
ax=plt.axes(projection='3d')
ax.scatter3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(45,215)
```
