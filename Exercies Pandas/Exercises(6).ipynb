### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",delimiter='\t')
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].replace('[\$, ]', '', regex=True).astype('float')
chipo.head()
```

```python
product_price=chipo[chipo['item_price']>10.00]
product_price
```

```python
product_name=product_price['item_name']
product_name_list=product_name.to_list()
result = set()
for item in product_name_list:
    result.add(item)

print("Unique product names:")
for unique_item in result:
    print(unique_item)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
filter_data=chipo.drop_duplicates(['item_name','quantity'])
one_quantity=filter_data[filter_data['quantity']==1]
price_item=one_quantity[['item_name','item_price']]
price_item
```
