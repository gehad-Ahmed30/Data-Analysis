### Step 1. Import the necessary libraries

```python
import pandas as pd  
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv). 

### Step 3. Assign it to a variable called 

```python
called=pd.read_csv(r"https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv")
called.head()
```

### Step 4.  What is the frequency of the dataset?

```python
called['Date']=pd.to_datetime(called['Date'])
date_diff=called['Date'].diff().dropna()
print('Intervals between dates:')
print(date_diff.value_counts())         # weekly data
```

### Step 5. Set the column Date as the index.

```python
called.set_index('Date',inplace=True)
called.head()
```

### Step 6. What is the type of the index?

```python
type(called.index)
```

### Step 7.  Change the frequency to monthly, sum the values and assign it to monthly.

```python
monthly=called.resample('M').sum()
monthly
```

### Step 8. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.

```python
monthly = monthly.dropna()
monthly
```

### Step 9. Good, now we have the monthly data. Now change the frequency to year.

```python
Year=monthly.resample('Y').sum()
Year
```
